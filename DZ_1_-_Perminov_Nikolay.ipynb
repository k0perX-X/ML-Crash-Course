```python
def print_fib(first, second, end):
    print(first, end=', ')
    while second <= end:
        print(second, end=', ')
        second, first = first + second, second
```

```python
print_fib(1,1,8)
```

```python
from random import randint
```

```python
a = []
for i in range(randint(1,100)):
    a.append(randint(-100000,100000))
```

```python
a
```

```python
for i in range(len(a)):
    for j in range(0, len(a) - 1):
        if a[j] < a[j+1]:
            a[j], a[j+1] = a[j+1], a[j]
```
